```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
data = pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

***Drop some unnecessary data***

```python
data= data.drop(columns=['RowNumber','CustomerId','Surname'],axis = 1)
```

```python
data.shape
```

***encoding catagorical variables***

```python
le_gender = LabelEncoder()
data['Gender'] = le_gender.fit_transform(data['Gender'])
data.head()
```

```python
data['Geography'].value_counts()
```

```python
ohe_geography = OneHotEncoder(drop='first')
geo_encoded = ohe_geography.fit_transform(data[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=ohe_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
data = pd.concat([data.drop(columns=['Geography']),geo_encoded_df],axis=1)
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
exited_counts = data['Exited'].value_counts()

plt.figure(figsize=(6,4))

plt.pie(exited_counts,shadow=True, explode=(0, 0.1),labels = exited_counts.index,autopct='%1.1f%%',startangle=90)
plt.title("Distribution of Exited Customers")
plt.show()
```

```python
data.corr()
```

```python
data.describe()
```

***Splitting Features and target***

```python
x = data.drop(columns=['Exited'],axis=1)
y = data[['Exited']]
```

```python
x.head()
```

```python
y.head()
```

***Split dataset into 70% and 30% test***

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=11)
```

```python
x_train.shape
```

```python
x_test.shape
```

***Decision Tree***

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
