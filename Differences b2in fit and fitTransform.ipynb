```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
data = np.array([[1,2,3,4],[5,1,2,3],[9,8,3,1]])
```

```python
scaler = MinMaxScaler()
```

```python
data_transformed = scaler.fit_transform(data) #only by fit_transformed
```

***now fit and transform step by step***

```python
scaler.fit(data)
data_transformed1 = scaler.transform(data)
```

```python
print("Trough fit_transformed:\n")
print(data_transformed)
print("\nThrough fit and transform seperately:\n")
print(data_transformed1)
```

***it seems same through fit_transform or fit and transform seperately***
