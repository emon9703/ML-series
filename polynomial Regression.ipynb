```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('icecream_sales.csv')
```

```python
data.head()
```

```python
x = data['Temperature']
y = data['Ice Cream Sales']
```

```python
plt.scatter(x,y)
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Sales')
plt.title('Ice Cream Sales prediction')
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x.values.reshape(-1,1),y)
```

```python
sales_pred = reg.predict(x.values.reshape(-1,1))
```

```python
reg.predict([[-4.662263]])
```

***Non-Linear Relationship***

```python
plt.scatter(x,y)
plt.plot(x,sales_pred,'r')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Sales')
plt.title('Ice Cream Sales prediction')
```

```python
reg.score(x.values.reshape(-1,1),y)
```

***Polynomial Relation***

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
```

```python
x_poly = poly.fit_transform(x.values.reshape(-1,1))
```

```python
reg.fit(x_poly,y)
```

```python
poly_pred = reg.predict(x_poly)
```

```python
plt.scatter(x,y)
plt.plot(x,poly_pred,'r')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Sales')
plt.title('Ice Cream Sales prediction')
```

```python
reg.score(x_poly,y)
```

```python
# Assuming you have already fitted the model as shown in the previous steps
temperature = np.array([[-4.662263]])
temperature_poly = poly.transform(temperature)
predicted_sales = reg.predict(temperature_poly)

print("Predicted Ice Cream Sales:", predicted_sales[0])
```
