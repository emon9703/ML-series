```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,recall_score, precision_score,classification_report
```

```python
x, y = make_classification(n_samples=1000, n_features=5, n_classes=2, random_state=11)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=11)
```

```python
clf = DecisionTreeClassifier(random_state=11)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

***calculate the confusion metrix***

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=['Predicted Negative','Predicted positive'],yticklabels=['Actual negative','Actual positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion metris heatmap')
```

```python
tn, fp, fn, tp = cm.ravel()
```

```python
print(f"True negative:{tn}")
print(f"False Positive:{fp}")
print(f"False negative:{fn}")
print(f"True Positive:{tp}")
```

```python
accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
```

```python
print(f"accuracy: {accuracy}")
print(f"precision: {precision}")
print(f"recall: {recall}")
print(f"f1: {f1}")
```
