```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('emails.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates('text')
```

```python
data
```

```python
data['spam'].value_counts()
```

```python
x = data['text']
```

```python
x
```

```python
y = data['spam']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=11)
```

***data preprocessing***

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
x_train = cv.fit_transform(xtrain)
```

```python
x_train.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
MNB = MultinomialNB()
```

```python
MNB.fit(x_train, ytrain)
```

```python
x_test = cv.transform(xtest)
```

```python
MNB.score(x_test, ytest)
```

```python
email = ['Hey emon how are you? can you please provide me a youtube channel', 'hey you win iphone 16 .please submit your personal information']
```

```python
cv_email = cv.transform(email)
```

```python
MNB.predict(cv_email)
```
