```python
x = data.drop(['Profit'],axis=1)
```

```python
x.head()
```

```python
y = data['Profit']
```

```python
y.head()
```

***one hot encoding for Area column***

```python
city = pd.get_dummies(x['Area'],drop_first=True,dtype=int)
```

```python
city.head()
```

```python
x = x.drop('Area',axis=1)
```

```python
x.head()
```

```python
x = pd.concat([x,city],axis=1)
```

```python
x.head()
```

```python
#Splitting the dataset into trainning and test set
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x , y , test_size=0.25, random_state=11)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(xtrain,ytrain)
```

```python
xtest
```

```python
ytest
```

```python
pred = regressor.predict(xtest)
```
