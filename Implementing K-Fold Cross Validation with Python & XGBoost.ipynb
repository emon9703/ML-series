```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, KFold, cross_val_score, StratifiedKFold
import warnings
warnings.filterwarnings('ignore')
```

```python
x, y = make_classification(n_samples=1000, n_features=20, n_informative=2, n_redundant=10, random_state=11)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=11)
```

***XGBoost Classifier***

```python
clf = xgb.XGBClassifier()
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

***K-Fold cross-validation***

```python
kf = KFold(n_splits=5, shuffle=True, random_state=11)
kf_score = cross_val_score(clf, x, y, cv=kf)
```

```python
kf_score
```

```python
kf_score.mean()
```

***StratifiedKFold***

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=11)
skf_score = cross_val_score(clf, x, y, cv=skf)
```

```python
skf_score
```

```python
skf_score.mean()
```
